# Assignment Name : AI Around Me
## Description: List 10 real-life AI applications you use daily and explain how ML might work behind each in 1–2 lines.

1. Instagram Feed Recommendations – My feed always shows posts I’m interested in. Machine learning analyzes my likes, shares, watch time, and interactions to predict what content I’ll engage with most.

2. Google Maps Navigation – It suggests the fastest route and shows traffic updates. ML models use real-time traffic data and past travel patterns to estimate congestion and travel time.

3. YouTube Video Suggestions – The recommended videos feel very personalized. ML tracks my viewing history, search queries, and watch duration to suggest similar content.

4. Spotify Music Recommendations – Discover Weekly playlists match my taste surprisingly well. ML analyzes my listening patterns and compares them with users who have similar preferences.

5. Gmail Spam Filter – Spam emails automatically go to a separate folder. ML models are trained on millions of emails to detect patterns common in spam messages.

6. Google Assistant / Voice Search – It understands my voice commands accurately. Speech recognition models convert voice to text using deep learning trained on large audio datasets.

7. Amazon Product Recommendations – It suggests products based on my searches. ML uses collaborative filtering to recommend items that similar users have viewed or purchased.

8. Face Unlock on My Phone – My phone unlocks instantly when I look at it. AI uses computer vision and facial recognition models to match my face with stored data.

9. Keyboard Autocorrect & Predictions – My keyboard predicts the next word while typing. ML models learn from my typing habits and common language patterns.

10. Netflix Movie Recommendations – It suggests movies based on what I’ve watched. ML analyzes my watch history and ratings to recommend similar genres or themes.
