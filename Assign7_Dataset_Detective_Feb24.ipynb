```python
# DATASET DETECTIVE

import pandas as pd

# Load dataset from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display top rows
print("Top 5 rows of the dataset:")
print(df.head())

# Highest value in numeric columns
print("\nHighest value in each numeric column:")
print(df.max(numeric_only=True))

# Count missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Statistics
print("\nBasic Statistics:")
print(df.describe())
```

Insights :

1. The Fare column has the highest value among numeric columns.

2. Some columns like Age and Cabin contain missing values.

3. The dataset contains both categorical and numerical data.

4. Passenger ages vary widely, showing a diverse age group.

5. Ticket fares vary significantly between passengers.
