```python
# KEYWORD EXTRACTOR


# NLP Mini App - Keyword Extractor

# Import required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Sample paragraph
text = """
Artificial Intelligence is transforming the world by enabling machines
to learn from data and make intelligent decisions. NLP is an important
branch of AI that helps computers understand human language.
"""

# Convert text to lowercase
text = text.lower()
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')


# Tokenize the words
words = word_tokenize(text)

# Remove punctuation and stopwords
stop_words = set(stopwords.words('english'))

filtered_words = []

for word in words:
    if word not in stop_words and word not in string.punctuation:
        filtered_words.append(word)

# Count word frequency
word_freq = Counter(filtered_words)

# Display top keywords
print("Extracted Keywords:\n")

for word, freq in word_freq.most_common(10):
    print(word, ":", freq)
```
