Five Uses of Face/Object Detection
1. Face Unlock in Smartphones :

Face detection is used in smartphones for secure unlocking systems. The camera detects and recognizes the user’s face before granting access to the device. This improves security and provides quick authentication.

2. Attendance Management System :

Object and face detection can be used in schools, colleges, and offices to automatically mark attendance. The system detects faces from a camera feed and records attendance without manual intervention.

3. Traffic Monitoring System :

Object detection helps identify vehicles, pedestrians, and traffic signals on roads. It is used in smart traffic systems to reduce congestion, monitor rule violations, and improve road safety.

4. Security and Surveillance :

Face detection is commonly used in CCTV surveillance systems. Suspicious persons or unauthorized individuals can be detected automatically, helping improve security in public places, airports, and banks.

5. Medical and Healthcare Applications :

Object detection can assist doctors in analyzing medical scans such as X-rays or MRI images. It helps identify abnormalities, tumors, or diseases more efficiently and accurately.

```python
# Face Detection using OpenCV
# Smart Attendance System Basic Demo

# Import libraries
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Upload image
uploaded = files.upload()

# Get uploaded image name
image_path = list(uploaded.keys())[0]

# Load image
image = cv2.imread(image_path)

# Convert image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load Haar Cascade face detector
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30,30)
)

# Print number of faces detected
print("Number of Faces Detected:", len(faces))

# Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (255,0,0), 2)

# Display image
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.title("Face Detection Output")
plt.axis('off')
plt.show()
```
