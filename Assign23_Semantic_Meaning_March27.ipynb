```python
# Install spaCy
!pip install spacy

# Download model
!python -m spacy download en_core_web_sm

# Import library
import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Word pairs
word_pairs = [
    ("car", "automobile"),
    ("teacher", "educator"),
    ("happy", "joyful"),
    ("big", "large"),
    ("student", "learner")
]

# Calculate similarity
for word1, word2 in word_pairs:
    token1 = nlp(word1)
    token2 = nlp(word2)

    similarity = token1.similarity(token2)

    print(f"{word1} - {word2}")
    print("Similarity Score:", similarity)
    print()
```

Explain Semantic Similarity.

Semantic similarity refers to the degree to which two words, phrases, or sentences are related in meaning. It helps computers understand language by identifying words that convey similar ideas even if they are written differently. For example, words like “car” and “automobile” or “happy” and “joyful” are considered semantically similar because they share nearly the same meaning. In Natural Language Processing (NLP), semantic similarity plays an important role in applications such as search engines, chatbots, recommendation systems, text summarization, and language translation. AI systems use different techniques such as word embeddings, cosine similarity, and transformer models to measure semantic similarity by analyzing the context and relationships between words. This concept improves the ability of machines to understand human language more accurately and provide relevant responses.
