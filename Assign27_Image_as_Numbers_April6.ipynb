```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Load image
image = cv2.imread(image_path)

# Check image loaded properly
if image is None:
    print("Error: Unable to load image")
else:
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Print image details
    print("Image Shape:", image.shape)
    print("Total Pixels:", image.size)
    print("Data Type:", image.dtype)

    # Print pixel values
    print("Pixel Value at (100,100):", image[100,100])

    # Print channel values
    print("Blue Channel:", image[100,100,0])
    print("Green Channel:", image[100,100,1])
    print("Red Channel:", image[100,100,2])

    # Display image
    plt.imshow(image_rgb)
    plt.title("Uploaded Image")
    plt.axis('off')
    plt.show()
```

## Explanation :

An image may look like a normal photograph to us, but for a computer it is simply a collection of numbers arranged in rows and columns. Every tiny dot in an image is called a pixel, and each pixel stores numerical values that represent colors. By using Python and OpenCV, we can read these values and understand how images are handled internally by computers.

In this program, the image is first uploaded and loaded into Python. When we print the shape of the image, it tells us the dimensions of the picture. For example, an output like (500, 700, 3) means the image has a height of 500 pixels, a width of 700 pixels, and 3 color channels. These three channels represent the colors Red, Green, and Blue (RGB) which combine together to create the final image we see.

The program also prints the pixel values at a particular position in the image. A pixel value is a set of numbers that determines the intensity of colors at that location. For example, [120, 180, 200] means the pixel contains different intensities of blue, green, and red colors. The values usually range from 0 to 255, where 0 represents no intensity and 255 represents maximum intensity.

The channels of the image are explained separately as:
*   Blue Channel → controls the amount of blue color
*   Green Channel → controls the amount of green color
*   Red Channel → controls the amount of red color


By combining these three channels in different intensities, computers can generate millions of colors. This assignment helps in understanding that images are not just pictures, but numerical data that can be processed, analyzed, and modified using programming techniques.
