```python
# STORY TELLING WITH GRAPH

import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "Class": ["First", "Second", "Third"],
    "Passengers": [216, 184, 491],
    "Survived": [136, 87, 119]
}

df = pd.DataFrame(data)

# Bar Chart
plt.figure()
plt.bar(df["Class"], df["Passengers"])
plt.title("Number of Passengers by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()

# Pie Chart
plt.figure()
plt.pie(df["Passengers"], labels=df["Class"], autopct='%1.1f%%')
plt.title("Passenger Distribution by Class")
plt.show()

# Histogram (using survival numbers)
plt.figure()
plt.hist(df["Survived"], bins=5)
plt.title("Distribution of Survivors")
plt.xlabel("Number of Survivors")
plt.ylabel("Frequency")
plt.show()
```

# Short Data Story :

1. The graphs help us understand the passenger data in a simple and clear way.
From the bar chart, we can see that the third class had the highest number of passengers compared to the first and second classes. This shows that most people traveling on the ship chose the lower ticket category.

2. The pie chart also shows a similar trend. A large portion of the passengers belonged to the third class, while first class had the smallest share. This indicates that only a small number of people could afford luxury travel at that time.

3. The histogram of survivors gives an idea about how survival numbers were distributed. Even though third class had the most passengers, the number of survivors from that class was not equally high. This suggests that passenger class might have affected the chances of survival, possibly due to differences in access to safety resources.
