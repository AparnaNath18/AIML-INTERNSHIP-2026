# WHY PREPROCESSING IS NEEDED?

1. Removes noise like emojis, punctuation, and special characters
2. Converts slang and abbreviations into standard words
3. Corrects spelling mistakes and typos
4. Converts text into a consistent format (e.g., lowercase)
5. Removes unnecessary words (stopwords)
6. Reduces data complexity
7. Makes text easier for machines to understand
8. Improves the accuracy and performance of models
