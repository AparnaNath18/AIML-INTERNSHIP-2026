```python
# CUSTOMER SEGMENTATION

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mall_customers.csv"

try:
    df = pd.read_csv(url)
except:
    data = {
        "Annual Income (k$)": [15,16,17,18,19,20,60,62,65,68,70,72],
        "Spending Score (1-100)": [39,40,42,43,45,46,55,57,60,62,65,67]
    }
    df = pd.DataFrame(data)

print("Top rows of dataset:")
print(df.head())

X = df[["Annual Income (k$)", "Spending Score (1-100)"]]

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

print("\nDataset with clusters:")
print(df.head())

plt.figure()
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=df["Cluster"])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation using K-Means")
plt.show()
```

*   Cluster 0 → Medium income & medium spending
*   Cluster 1 → High income & high spending
*   Cluster 2 → Low income & low spending
