```python
# DATA DOCTOR

import pandas as pd
import numpy as np

# Creating the dataset
data = {
    "Name": ["Aparna", "Rahul", "Sneha", "Rahul", "Arjun", "Meera", "Kiran"],
    "Age": [20, 21, 19, 21, np.nan, 20, 22],
    "Department": ["Data Science", "data science", "AI", "data science", "AI", "Ai", "Data science"],
    "Marks": [85, 78, 90, 78, 88, 92, np.nan]
}

df = pd.DataFrame(data)

# Show uncleaned dataset
print("Uncleaned Dataset:")
print(df)

# Data Cleaning Steps

# Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

# Remove duplicate rows
df = df.drop_duplicates()

# Standardize text
df["Department"] = df["Department"].str.lower()

# Show cleaned dataset
print("\nCleaned Dataset:")
print(df)
```

# Why Data Cleaning matters?
Data cleaning is important because real-world datasets often contain errors such as missing values, duplicates, and inconsistent formatting. If these issues are not handled properly, they can lead to incorrect analysis and poor machine learning model performance. By cleaning the data, we ensure that the dataset becomes accurate, consistent, and reliable for further analysis and decision-making.
