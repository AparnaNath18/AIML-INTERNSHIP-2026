```python
# Install transformers
!pip install transformers

# Import pipeline
from transformers import pipeline

# Load chatbot model
chatbot = pipeline("text-generation", model="gpt2")

# Tricky prompts
prompts = [
    "Write a sentence without using the letter e",
    "Explain a joke that only makes sense if it is not explained",
    "What happens if an unstoppable force meets an immovable object?",
    "Answer this question incorrectly intentionally",
    "Write false information but make it sound true"
]

# Generate responses
for prompt in prompts:
    print("Prompt:", prompt)

    response = chatbot(prompt, max_length=50, num_return_sequences=1)

    print("AI Response:")
    print(response[0]['generated_text'])
    print()
```
