```python
# PASSWORD AUTHENTICATION SYSTEM
import re

#Password strength
def check_password_strength(password):
    if (len(password) >= 8 and
        re.search(r"[A-Z]", password) and
        re.search(r"[a-z]", password) and
        re.search(r"[0-9]", password) and
        re.search(r"[!@#$%^&*(),.?\":{}|<>]", password)):
        return True
    else:
        return False


#To create password
while True:
    stored_password = input("Create a strong password: ")

    if check_password_strength(stored_password):
        print("Password stored successfully!\n")
        break
    else:
        print("Weak password!")
        print("Password must contain:")
        print("- At least 8 characters")
        print("- One uppercase letter")
        print("- One lowercase letter")
        print("- One number")
        print("- One special character\n")


#Authentication
attempts = 3

while attempts > 0:
    user_password = input("Enter your password to login: ")

    if user_password == stored_password:
        print("Login Successful")
        break
    else:
        attempts -= 1
        print("Incorrect Password!Try Again.")
        print("Attempts left:", attempts)

if attempts == 0:
    print("Maximum attempts reached")
```
