```python
# NUMPY SPEED TEST

import time
import numpy as np

# Create 1 million numbers
size = 1_000_000

# Python List
start = time.time()
py_list = list(range(size))
py_result = [x * 2 for x in py_list]
end = time.time()
print("Python List Time:", end - start)

# NumPy Array
start = time.time()
np_array = np.arange(size)
np_result = np_array * 2
end = time.time()
print("NumPy Array Time:", end - start)
```

Observations :

1. NumPy arrays performed the operation much faster than Python lists when working with one million numbers.
2. Python lists process elements one by one, which increases execution time.
3. NumPy uses optimized operations, making it more efficient for large numerical datasets.
