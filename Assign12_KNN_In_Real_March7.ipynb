```python
# KNN IN REAL LIFE (Small similarity example)

from sklearn.neighbors import NearestNeighbors
import numpy as np

# User ratings dataset
# Columns: Action, Comedy, Drama
data = np.array([
    [5, 3, 2],  # User A
    [4, 3, 1],  # User B
    [1, 5, 4]   # User C
])

# Apply KNN
knn = NearestNeighbors(n_neighbors=2, metric='euclidean')
knn.fit(data)

# Find neighbors for User A
distances, indices = knn.kneighbors([data[0]])

print("Nearest Neighbors of User A:", indices)
print("Distances:", distances)
```

# **Explanation (Netflix-like Recommendation using KNN):**

The **K-Nearest Neighbors (KNN)** algorithm can be used in recommendation systems similar to how the Netflix platform suggests movies or shows to its users. Instead of making complex predictions, KNN simply looks for people who have similar tastes. It compares the viewing or rating patterns of different users and finds those who are most similar to each other.
For instance, if two users have watched many of the same movies and have given similar ratings, the system considers them to have similar preferences. If one of those users watches a new movie and likes it, the system may recommend that movie to the other user as well. This works because people with similar tastes often enjoy similar content.
In this way, KNN helps recommendation systems understand user behavior and suggest movies or shows that a person is more likely to enjoy, making the experience more personalized and engaging.
