## Mini Project Proposal

### Title:

Facial Expression Recognition using CNN

---

### Problem Statement:

Understanding human emotions is important in areas like human-computer interaction, mental health analysis, and smart applications. However, machines cannot naturally interpret facial expressions. This project aims to develop a system that can automatically detect and classify human emotions from facial images.

---

### Dataset:

The project uses the FER-2013 dataset. It contains grayscale images of faces categorized into emotions such as Happy, Sad, Angry, Surprise, and Neutral.

---

### Algorithm Used:

The model will use a Convolutional Neural Network (CNN).

* Suitable for image data
* Automatically extracts features
* Provides good accuracy for classification tasks

---

### Methodology:

1. Load and preprocess the image dataset
2. Normalize pixel values and reshape images
3. Build a CNN model with convolution and pooling layers
4. Train the model using training data
5. Evaluate the model using test data
6. Predict emotions from new images

---

### Expected Output:

* The model classifies facial expressions into different emotion categories
* Displays predicted emotion labels such as Happy, Sad, etc.
