```python
# HOUSE PRICE PREDICTOR

import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([[500], [800], [1000], [1200], [1500]])
y = np.array([100, 150, 200, 230, 300])
model = LinearRegression()
model.fit(X, y)
print("Predicted Prices for Training Data:")
for i in range(len(X)):
    predicted_price = model.predict([X[i]])
    print("House size:", X[i][0], "sq ft -> Predicted price:", round(predicted_price[0],2), "thousand")
size = int(input("\nEnter a new house size in sq ft: "))
predicted_price = model.predict([[size]])
print("Predicted price for", size, "sq ft house:", round(predicted_price[0],2), "thousand")
```
