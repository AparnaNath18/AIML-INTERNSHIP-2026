```python
# Weak and strong prompts

prompts = {
    "Resume": {
        "Weak": "Write a resume",
        "Strong": "Create a professional resume for a Data Science student with skills in Python, Machine Learning, and NLP"
    },

    "Business Idea": {
        "Weak": "Give a business idea",
        "Strong": "Suggest a low-investment AI startup idea for students in the education sector"
    },

    "Study Plan": {
        "Weak": "Make a study plan",
        "Strong": "Create a 30-day study plan for learning Python and Machine Learning basics for beginners"
    },

    "Email": {
        "Weak": "Write an email",
        "Strong": "Write a formal email requesting internship opportunities in the AI field"
    }
}

# Display prompts
for category, values in prompts.items():
    print(f"\n{category}")
    print("Weak Prompt :", values["Weak"])
    print("Strong Prompt :", values["Strong"])
```
